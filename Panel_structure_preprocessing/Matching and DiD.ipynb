```python
import pandas as pd
import pymatch
from pymatch.Matcher import Matcher
import numpy as np
```

```python
df =  pd.read_csv(r"covid_spending_cases_mobility_panel.csv")
dfStates = pd.read_csv(r"../us-state-ansi-fips.csv")
dfStates = dfStates[["stname", " st", "is_red_state"]]
dfStates.columns = ["stname", "statefips", "is_red_state"]
```

```python
df = df.merge(dfStates[["stname", "statefips", "is_red_state"]], on = "statefips")
```

Data to wide format for matching procedures

```python
df_wide = df.loc[df["year"] == 2020].pivot(index = "statefips", columns = "week", values = "case_rate")
```

```python
df_wide.columns = ["cases_week_" + str(n) for n in df_wide.columns]
```

```python
df_wide.columns
```

```python
#df_wide.columns = ['_'.join(str(s).strip() for s in col if s) for col in df_wide.columns]
```

```python
df_treatment_assignment = df.groupby("statefips").agg(stayathome_advisory = ("stayathome_advisory", max))
```

```python
df_wide = df_wide.merge(dfStates[["stname", "statefips", "is_red_state"]], on = "statefips")
df_wide = df_wide.merge(df_treatment_assignment, on = "statefips")
```

# Matching

```python
#select relevant columns
mask = ['statefips', 'cases_week_10',
       'cases_week_11', 'cases_week_12', 'cases_week_13', 'cases_week_14',
       'cases_week_15', 'cases_week_16', 'cases_week_17', 'cases_week_18',
       'cases_week_19','is_red_state',
       'stayathome_advisory']

df_wide = df_wide[mask]
```

```python
test = df_wide.loc[df_wide.stayathome_advisory == 1]
control = df_wide.loc[df_wide.stayathome_advisory == 0]
test.loc[:,'stayathome_advisory'] = 1
control.loc[:,'stayathome_advisory'] = 0
```

```python
m = Matcher(test, control, yvar="stayathome_advisory", exclude=["statefips"])
```

```python
np.random.seed(10)
m.fit_scores(balance=False, nmodels=100)
```

```python
m.predict_scores()
m.plot_scores()
```

```python
m.tune_threshold(method='random')
```

```python
m.match(method="min", nmatches=1, threshold=0.0001)
m.record_frequency()
```
